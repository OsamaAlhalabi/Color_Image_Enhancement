# Color Image Enhancement using Fuzzy Logic
**Note** this notebook just for testing

```python
import cv2 as cv
from matplotlib import pyplot as plt
from image_enhancement import ImageEnhancement
```

```python
def _plot_imgs(original_image, enhanced_image):
    plt.subplot(1, 2, 1)
    plt.imshow(original_image)
    plt.subplot(1, 2, 2)
    plt.imshow(enhanced_image)
    plt.show()
```

## 1. Apply the Enhancement on images in Gray Scale

```python
img = cv.imread('images/sample1.PNG')
gray_img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
```

```python
img_enh = ImageEnhancement(gray_img, 2, 2, 1, False)
```

```python
result = img_enh.enhance_gray_img()
```

```python
_plot_imgs(img, result)
```

## 2. Apply the Enhancement on colored images

```python
img = cv.imread('images/sample3.png')
img = cv.resize(img,(300, 300), interpolation = cv.INTER_AREA)
img_enh = ImageEnhancement(img, 2, 2, 1, True)
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
result = img_enh.enhance_colored_img()
```

```python
_plot_imgs(img, cv.cvtColor(result, cv.COLOR_BGR2RGB))
```
